```python
import numpy as np
image3 = open('download.jpg')     
print(image3)
```

```python
print( "input image")
ROWS = 1080    
COLS =  980
img = np.fromfile(image3, dtype = np.uint8, count = ROWS * COLS)
print("Dimension :", img.ndim)
print("Size: ", img.size)
```

```python
img.shape = (img.size // COLS,COLS)
print("New dimension of the array:", img.ndim)
print("----------------------------------------------------")
print(" The 2D array of the original image is: \n", img)
print("----------------------------------------------------")
print("The shape of the original image array is: ", img.shape)
```

```python
print("Saving the output image")
img.astype('int8').tofile('NewImage.raw')
print("File successfully saved")

image3.close()
```
