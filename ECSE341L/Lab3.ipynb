# Q4
<strong> Negative  of  an  image</strong>

```python
import cv2 
import matplotlib.pyplot as plt 
  
   
img1 = cv2.imread('9.jpeg', 1) 
plt.imshow(img1) 
plt.show() 
 
  
height, width, _ = img1.shape 
  
for i in range(0, height - 1): 
    for j in range(0, width - 1): 
          
        pixel = img1[i, j]  
        pixel[0] = 255 - pixel[0] 
        pixel[1] = 255 - pixel[1]  
        pixel[2] = 255 - pixel[2] 
          
        img1[i, j] = pixel 
  

plt.imshow(img1) 
plt.show() 
```

# Q5
<strong>Creating a circular mask on the input image </strong>

```python
import cv2
import numpy as np
img = cv2.imread('9.jpeg')
mask = np.zeros(img.shape, dtype = np.uint8)
mask = cv2.circle(mask, (260, 300), 225, (255, 255, 255), -1)

result = cv2.bitwise_and(img, mask)
result[mask == 0] == 255
```

```python
plt.subplot(2, 2, 1)
plt.imshow(img)

plt.subplot(2, 2, 2)
plt.imshow(result)
```

# Q6
<strong>Reading and displaying multiple images at once</strong>

```python
import cv2
import glob
path = "data\*.*"
for file in glob.glob(path):
    print(file)
    a= cv2.imread(file)
    print(a)
    c = cv2.cvtColor(a, cv2.COLOR_BGR2RGB)
    cv2.imshow('Color image', c)
    plt.imshow(c)
    #wait for 1 second
    k = cv2.waitKey(0)
   
    cv2.destroyAllWindows()
```
